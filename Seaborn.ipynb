## Part 1: Seaborn Anatomy (Figure-Level vs. Axes-Level)

Seaborn divides its tools into two distinct architectural components:
* **Axes-Level Functions:** These functions draw directly onto an existing Matplotlib plot window object. Examples include `sns.scatterplot()` or `sns.histplot()`. They provide precision control when building multi-chart layouts.
* **Figure-Level Functions:** These functions control the entire overall window grid container themselves. Examples include `sns.relplot()` or `sns.catplot()`. They are highly automated and excel at generating multi-panel comparisons seamlessly.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Let's confirm everything is imported successfully
print("Seaborn version:", sns.__version__)
```

## Part 2: Usage in Notebooks

When working with Seaborn inside a notebook, you can set a global visual theme using `sns.set_theme()`. This alters all subsequent chart displays—applying clean background grids, modern text formatting, and balanced color palettes automatically without manual inline customization.


```python
# Apply the default Seaborn aesthetic configurations globally
sns.set_theme(style="darkgrid")
```

## Part 3: Relational Plots (Scatter & Line plots)

Let's construct a sample dataset tracking advertising spending against product sales rows using a Pandas DataFrame.


```python
ad_data = {
    "Spend_USD": [100, 250, 400, 600, 750, 900, 1100, 1300],
    "Sales_Units": [12, 35, 42, 68, 70, 89, 95, 120],
    "Platform": ["Social", "Search", "Social", "Search", "Social", "Search", "Social", "Search"]
}
df_ads = pd.DataFrame(ad_data)
```

We can now pass our DataFrame directly into `sns.scatterplot()`. We simply declare the column names for `x` and `y`, and identify our source dataset inside the `data` parameter. The `hue` parameter automatically color-codes points by category text.


```python
# Create a relational scatter plot color-coded by platform type
plt.figure(figsize=(7, 4))
sns.scatterplot(x="Spend_USD", y="Sales_Units", hue="Platform", data=df_ads, s=100)
plt.title("Advertising Performance Assessment")
plt.show()
```

## Part 4: Distribution Plots (Histograms)

Seaborn's `sns.histplot()` simplifies creating histograms by automatically computing balanced interval divisions and adding a smoothing curve line option (`kde=True`) to show the shape of the dataset.


```python
# Generate simulated continuous measurement values
np.random.seed(42)
sample_scores = np.random.normal(loc=75, scale=10, size=300)

# Plot distribution density
plt.figure(figsize=(7, 4))
sns.histplot(sample_scores, kde=True, color="teal")
plt.title("Observed Measurement Spread")
plt.show()
```

## Part 5: Categorical Plots (Bar plots)

`sns.barplot()` cuts out manual data prep by taking raw un-aggregated lists, calculating grouping averages automatically, and generating structural confidence interval lines on top of the bars.


```python
# Plot category averages directly from the original DataFrame columns
plt.figure(figsize=(6, 4))
sns.barplot(x="Platform", y="Sales_Units", data=df_ads, palette="muted")
plt.title("Average Output Units by Channel Type")
plt.show()
```

## Part 6: Heatmaps

Seaborn's `sns.heatmap()` layers quantitative grid numbers across colored matrices. It pairs perfectly with the pandas `.corr()` matrix function to find relationships between data table features.


```python
# Compute a standard numerical relationship matrix block
corr_matrix = df_ads.corr(numeric_only=True)

# Render using an interactive annotation grid scale
plt.figure(figsize=(5, 3.5))
sns.heatmap(corr_matrix, annot=True, cmap="Blues", vmin=-1, vmax=1)
plt.title("Feature Attribute Matrix Correlation")
plt.show()
```
