```.arange(start,stop,step)``` Generates a 1D array with evenly spaced values based on your parameters.

```start```: The first value in the array (inclusive).

```stop```: The end boundary (exclusive—the array stops before reaching this value).

```step```: The spacing between each consecutive value.

````python
np.arange(1,20,5)
````

```.linspace(start,stop,step)``` Generates a 1D array of num evenly spaced samples over a specified interval.

```start```: The first value in the sequence.

```stop```: The final value (inclusive—unlike arange, the endpoint is included).

```n```: The number of samples to generate (defaults to 50).

````python
np.linspace(1,20,5)
````

## Broadcasting and vectorization

Broadcasting allows NumPy to perform arithmetic on arrays of different shapes by "stretching" the smaller array to match the larger one.

````python
a = np.array([1, 2, 3])
b = 2
print(a,"*",b,"=",a * b)                # [2, 4, 6] (Broadcasting scalar 2)
````

````python

mat = np.array([[1], [2]])
print(mat,"+",a,"=",mat + a)              # Adds [1,2,3] to both rows [1] and [2]
````

## Memory Managing

Crucial for avoiding bugs where modifying a slice unexpectedly changes the original array.

- **View**: Slicing creates a view. Modifying the slice modifies the original.

- **Copy**: ```.copy()``` creates a new object. Modifying it does not touch the original.

````python
arr = np.array([1, 2, 3])
view = arr[0:2]             # This is a view
view[0] = 99
print(arr)
````

````python
arr = np.array([1, 2, 3])
copy = arr[0:2].copy()      # This is a copy
copy[0] = 5
print(arr)
````
