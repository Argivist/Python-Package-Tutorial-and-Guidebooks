## Part 1: Pandas Anatomy

Before manipulating data, you need to understand the two main layers of Pandas:
* **Series:** A single column of data. It is a one-dimensional array where every entry has an index label.
* **DataFrame:** A full table of data. It is a two-dimensional structure made by combining multiple Series columns side-by-side. It has row index labels and column text headers.


```python
import pandas as pd
import numpy as np

# Let's import the library and confirm it is ready
print("Pandas version:", pd.__version__)
```

## Part 3: Creating Data From Scratch

Let's create a functional DataFrame by combining python lists inside a dictionary structure. Each key becomes a column header, and each list becomes the column's data values.


```python
store_inventory = {
    "Item_ID": [101, 102, 103, 104, 105],
    "Item_Name": ["Keyboard", "Desk Lamp", "Headphones", "USB Cable", "Webcam"],
    "Category": ["Electronics", "Furniture", "Electronics", "Accessories", "Electronics"],
    "Price": [45.00, 32.50, 89.99, 12.00, 65.00],
    "Units_Sold": [120, 45, 85, 310, 60]
}

df = pd.DataFrame(store_inventory)
```

## Part 4: Inspecting Data

When loading data for the first time, you should always inspect its shape and view the first few rows to make sure it loaded correctly. We use the `.head(n)` method to view the top `n` rows.


```python
# View the first 3 rows of the table
df.head(3)
```

We can also check the data types of each column and the overall count of non-empty entries using `.info()`.


```python
# Check column data types and memory size
df.info()
```

## Part 5: Filtering & Selection

To pull out a single column from our table, pass the column's text name inside square brackets. This returns a Pandas Series object.


```python
# Isolate the Price column
df["Price"]
```

To filter rows based on a specific condition, we pass a logical statement inside the brackets. This checks every row and only keeps the ones that evaluate to `True`.


```python
# Only keep rows where the Category equals 'Electronics'
electronics_df = df[df["Category"] == "Electronics"]
electronics_df
```

## Part 6: Basic Aggregation (Groupby)

To find patterns across data categories, we use the `groupby()` method. It groups identical items together, allowing us to compute averages or sums for each distinct group.


```python
# Group rows by Category and find the average Price and Units_Sold for each
summary_df = df.groupby("Category")[["Price", "Units_Sold"]].mean()
summary_df
```

## Part 7: Converting Pandas to NumPy Guide

Sometimes you need to pass your data into a machine learning library or a mathematical tool that only accepts raw numbers. Since Pandas is built on top of NumPy, you can instantly strip away the row labels and column headers to turn your tables into raw NumPy arrays using the `.to_numpy()` method.


### Converting a Single Column (Series)

Let's start by seeing how to convert a single column (a Pandas Series) into a one-dimensional NumPy array.


```python
import pandas as pd
import numpy as np

# Create a single Pandas column (Series)
ages_series = pd.Series([22, 35, 58, 19, 45])

# Convert the single column into a raw NumPy array
ages_array = ages_series.to_numpy()
```

We can print out the result and check its type to prove that it is no longer a Pandas object, but a standard NumPy `ndarray`.


```python
# Display the new array and confirm its type
print("NumPy Array:", ages_array)
print("Data Type:", type(ages_array))
```

### Converting a Full Table (DataFrame)

You can do the exact same thing with a full table (a DataFrame). When you convert a full table, it becomes a two-dimensional NumPy grid of numbers. 

Let's create a quick table with two numerical columns.


```python
# Create a small DataFrame table
data_table = pd.DataFrame({
    "Height_cm": [165, 180, 172],
    "Weight_kg": [62, 78, 70]
})
data_table
```

```python


# Convert the entire table into a 2D NumPy grid
grid_array = data_table.to_numpy()
```

When we print out this grid array, you will see that the column names (`Height_cm`, `Weight_kg`) and row index numbers (`0, 1, 2`) are completely gone, leaving only the raw matrix of values.


```python
# Display the 2D grid array
print(grid_array)
```
